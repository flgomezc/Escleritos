```python
data = loadtxt("adultos_pesos.csv", usecols=(0,1,2,3,4,5,6),skiprows=2, delimiter=",")
```

```python
Replica=data[:,2]
Delta1 = data[:,4]-data[:,3]
Delta2 = data[:,5]-data[:,4]
Delta3 = data[:,6]-data[:,5]
DeltaT = data[:,6]-data[:,3]
```

```python
Delta1.shape
```

```python
index = []
listaPH = [7.0, 7.5, 8.05, 8.06]
replica = 1 ### Valores aceptados: 0=negro, 1=rojo

for PH in listaPH:
    index.append(where( (data[:,0]==PH) & (Replica==replica) )[0])

DeltaTPH_7_00  = DeltaT[index[0]]
DeltaTPH_7_50  = DeltaT[index[1]]
DeltaTPH_8_05  = DeltaT[index[2]]
DeltaTPH_8_05C = DeltaT[index[3]]

DeltaT_lista = [DeltaTPH_7_00, DeltaTPH_7_50, DeltaTPH_8_05, DeltaTPH_8_05C]

fig = plt.figure(1, figsize=(9, 6))
title(r"Cambio de peso Febrero-Mayo - Replica "+str(replica), fontsize=25)
ax = fig.add_subplot(111)
bp = ax.boxplot(DeltaT_lista, patch_artist=False)
ax.set_xticklabels(['7.00', '7.50', '8.05', "8.05-calceina"], fontsize=18)
ax.set_ylabel("$\Delta$ Peso (gramos)", fontsize=18)
ax.set_ylim(-1.5, 1.2 )
ax.axhline(y=0,linewidth=0.5, color='k')
savefig("delta_peso_total_replica"+str(replica)+".png")
savefig("delta_peso_total_replica"+str(replica)+".pdf")
```

```python
listaPH = [8.06]
for PH in listaPH:

    index1 = where( data[:,0]==PH )

    MAR = Delta1[index1]
    ABR = Delta2[index1]
    MAY = Delta3[index1]
    TOT = DeltaT[index1]

    data_to_plot = [ MAR, ABR, MAY, TOT]

    title=(str(PH))
    ax = fig.add_subplot(111)
    bp = ax.boxplot(data_to_plot, patch_artist=False)
    ax.set_xticklabels(['Feb-Mar', 'Mar-Abr', 'Abr-May', "$\Delta$ Total"], fontsize=18)
    ax.set_ylabel("$\Delta$ Peso (gramos)", fontsize=18)
    ax.set_ylim(-1.5, 1.1 )
    
    if (PH!=8.06):
        plt.text(1.9,0.9, "PH "+str(PH)    ,fontsize=20)
        savefig(str(PH)+".png")
    else:
        plt.text(1.9,0.9, "PH 8.05 Calceina",fontsize=20)
        savefig("8.05_calceina.png")
```
