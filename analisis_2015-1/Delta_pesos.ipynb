```python
data = loadtxt("adultos_pesos.csv", usecols=(0,1,2,3,4,5,6),skiprows=2, delimiter=",")
```

```python
Delta1 = data[:,4]-data[:,3]
Delta2 = data[:,5]-data[:,4]
Delta3 = data[:,6]-data[:,5]
DeltaT = data[:,6]-data[:,3]
```

```python
listaPH = [7.0]
for PH in listaPH:

    index1 = where( data[:,0]==PH )

    MAR = Delta1[index1]
    ABR = Delta2[index1]
    MAY = Delta3[index1]
    TOT = DeltaT[index1]

    data_to_plot = [ MAR, ABR, MAY, TOT]

    fig = plt.figure(1, figsize=(9, 6))
    title=(str(PH))
    ax = fig.add_subplot(111)
    bp = ax.boxplot(data_to_plot, patch_artist=False)
    ax.set_xticklabels(['Feb-Mar', 'Mar-Abr', 'Abr-May', "$\Delta$ Total"], fontsize=18)
    ax.set_ylabel("$\Delta$ Peso (gramos)", fontsize=18)
    ax.set_ylim()
    
    if (PH!=8.06):
        plt.text(1.9,0.9, "PH "+str(PH)    ,fontsize=20)
        savefig(str(PH)+".png")
    else:
        plt.text(1.9,0.9, "PH 8.05 Calceina",fontsize=20)
        savefig("8.05_calceina.png")
```
